```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head()
```

```python
df["Pclass"]. value_counts()
```

```python
df["Fare"].describe()
```

```python
# Allocate happiness to bins
binned = pd.cut(df["Fare"], bins = [0,7.91,14.45,31,512.3292], labels = ["Q1","Q2","Q3","Q4"])

# Add the binned values as a new categorical feature
df["Fare_band"] = binned
df
```

```python
df['Deck'] = df['Cabin'].str.extract('([A-Za-z]+)')
df["Deck"]. value_counts()
```

```python
import numpy as np
df["Deck"] = df["Deck"].replace(np.nan, "unknow")
df["Deck"]. value_counts()
```

```python
df["Survived"].groupby(df["Deck"])
```
